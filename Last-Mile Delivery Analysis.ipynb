```python
# Replacing null value
df['Instructions']=df['Instructions'].fillna('No Instructions')
```

```python
# Replacing null value
df['Discount construct'] = df['Discount construct'].fillna('No Discount')
```

```python
# Replacing null value by taking account of customer complaint
for i in df.index:

    if pd.isnull(df.loc[i, 'Rating']):

        # If complaint tag is null
        if pd.isnull(df.loc[i, 'Customer complaint tag']):

            df.loc[i, 'Rating'] = 5

        # Otherwise fill with restaurant average rating
        else:

            avg_rating = df[
                df['Restaurant ID'] == df.loc[i, 'Restaurant ID']
            ]['Rating'].mean()

            df.loc[i, 'Rating'] = round(avg_rating, 1)
```

```python
# Replacing null value
df['Review'] = df['Review'].fillna('No Command')
```

```python
# Replacing null value
df['Cancellation / Rejection reason'] = df['Cancellation / Rejection reason'].fillna('No cancellation')
```

```python
# Replacing null value
df['Restaurant compensation (Cancellation)'] = df['Restaurant compensation (Cancellation)'].fillna(0.0)
```

```python
# Replacing null value
df['Restaurant penalty (Rejection)'] = df['Restaurant penalty (Rejection)'].fillna(0.0)
```

```python
# Replacing null value
df['Customer complaint tag'] = df['Customer complaint tag'].fillna('No complain')
```

```python
# Replacing null value
df['Rider wait time (minutes)'] = df['Rider wait time (minutes)'].fillna(0.0)
```

```python
# Changing  the data type and replacing null value with logic
df['KPT duration (minutes)'] = pd.to_numeric(
    df['KPT duration (minutes)'],
    errors='coerce'
)

for i in df.index:
    if pd.isnull(df.loc[i, 'KPT duration (minutes)']):
        if df.loc[i, 'Cancellation / Rejection reason'] == 'Merchant device issue':

            avg_kpt = df[
                df['Restaurant ID'] == df.loc[i, 'Restaurant ID']
            ]['KPT duration (minutes)'].mean()

            df.loc[i, 'KPT duration (minutes)'] = round(avg_kpt, 2)
        else:
            df.loc[i, 'KPT duration (minutes)'] = 0.0
```

```python
#Useing IQR method for outlier in wait time
Q1 = df['Rider wait time (minutes)'].quantile(0.25)
Q3 = df['Rider wait time (minutes)'].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Create a new dataframe without outliers
df_no_outliers = df[
    (df['Rider wait time (minutes)'] >= lower_limit) &
    (df['Rider wait time (minutes)'] <= upper_limit)
]

print("Original rows:", len(df))
print("Rows after removing outliers:", len(df_no_outliers))
print("Outliers removed:", len(df) - len(df_no_outliers))
```
